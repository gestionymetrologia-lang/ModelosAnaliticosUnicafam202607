```python
# Ejercicio 5: Valores de Cadena

# 1. Declarar una cadena llamada saludo
saludo = "Hola, mundo!"

# 2. Mostrar el primer carácter de la cadena
print("Primer carácter:", saludo[0])

# 3. Mostrar los últimos 5 caracteres de la cadena
print("Últimos 5 caracteres:", saludo[-5:])

# 4. Concatenar otra cadena
saludo_completo = saludo + ", ¿cómo estás?"
print("Saludo completo:", saludo_completo)

# 5. Convertir la cadena a mayúsculas
saludo_mayusculas = saludo_completo.upper()
print("En mayúsculas:", saludo_mayusculas)

# 6. Reemplazar la palabra "mundo" por "Python"
saludo_reemplazo = saludo_completo.replace("mundo", "Python")
print("Reemplazo:", saludo_reemplazo)
```
