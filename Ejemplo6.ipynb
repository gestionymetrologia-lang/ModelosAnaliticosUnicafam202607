```python
# Una lista vacía
mi_lista=[]
# Lista con varios elementos
mi_lista=[1,2,3,"4", 5.0, True]
print(mi_lista[2])
```

```python
# Cambios en la lista
mi_lista[0]="Manuel"
print(mi_lista)
```

```python
# Agregar elementos a una lista
mi_lista.append("García")
print (mi_lista)
```

```python
# Agregar elementos a una posición específica
mi_lista.insert(3,"Tunja")
print(mi_lista)
```

```python
#Agregar varios elementos al final de la lista
mi_lista.extend([7,False])
print(mi_lista)
```

```python
# Quitar elementos
# eliminar objetos específicos
mi_lista.remove(True)
print(mi_lista)
```

```python
print("ACTUALIZACIÓN DE DATOS")
valor_buscar=input("Por favor ingrese el nombre para actualizar")
valor_reemplazar=input("Ingrese valor a reemplazar:")
mi_lista[mi_lista.index(valor_buscar)]=valor_reemplazar
print("Cambio ejecutado. Imprimiendo ejecutado")
print(mi_lista)
```
