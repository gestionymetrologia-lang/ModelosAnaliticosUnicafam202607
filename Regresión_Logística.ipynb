REGRESIÓN LOGÍSTICA

DIPLOMADO EN ANALÍTICA DE DATOS



```python
# Paso 1, Importar librerías

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Paso 2, Importar Set de Datos

dataset = pd.read_excel("/content/sample_data/Social_Network_Ads.xlsx")
```

```python
dataset
```

```python
# Paso 3. Separación de variables dependientes e independientes

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Paso 4. Separación datos de entrenamiento y datos de prueba

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
# Paso 5. Escalada de características

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
X = sc.fit_transform(X)
```

```python
X_train
```

```python

# Paso 6. Entrenamiento del modelo de regresión logística

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
# Paso 7. Ejemplo

print(classifier.predict(sc.transform([[32,150000]])))
```

```python
# Paso 8. Predicción Dataset de test

y_pred = classifier.predict(X)
dataset["Predicción"] = y_pred
```

```python
dataset
```

```python
# Paso 9. Matriz de Confusión

from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Paso 11. Gráfica Conjunto de Prueba

from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
