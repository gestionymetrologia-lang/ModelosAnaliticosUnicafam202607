```python
# EJERCICIO #2, SUMA ACUMULADA

suma_total = 0
suma_pares = 0

for numero in range(1, 101):

    suma_total = suma_total + numero

    if numero % 2 == 0:
        suma_pares = suma_pares + numero

print("Suma del 1 al 100:", suma_total)
print("Suma de los números pares:", suma_pares)
```

```python
# EJERCICIO #3. RECORRER UNA LISTA

precios = [12000, 45000, 8500, 67000, 3200]

numero_item = 1
articulos_mayores = 0

for precio in precios:
    print(f"Item {numero_item}: ${precio:,}")

    if precio > 10000:
        articulos_mayores = articulos_mayores + 1

    numero_item = numero_item + 1

print("Artículos que cuestan más de $10,000:", articulos_mayores)

```

```python
# EJERCICIO #4

saldo = 1000
retiro = 150
comision = 20
mes = 0

while saldo > 0:
    saldo = saldo - retiro - comision
    mes = mes + 1

    print(f"Mes {mes}: saldo = {saldo}")

print("Meses transcurridos:", mes)
```
