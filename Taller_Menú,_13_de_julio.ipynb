```python
# TALLER: MENÚ DE MI PERFIL CON DICCIONARIOS EN PYTHON
```

```python
# Creación del diccionario con datos personales
perfil = {
    "nombre": "Manuel Jacinto García Suárez",
    "ciudad": "Tunja",
    "edad": 59,
    "profesion": "Ingeniero Electricista",
    "universidad": "Universidad Nacional de Colombia",
    "hobbies": ["Lectura", "Juegos de Mesa", "Rompecabezas"],
    "lenguajes_programacion": "No",
    "correo": "gestionymetrologia@gmail.com"
}
```

```python
# Presentación del menú
```

```python
print("------------------------------------")
print("       MENÚ DE MI PERFIL")
print("------------------------------------")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato existente")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con pop")
print("5. Consultar un dato con get")
print("6. Ver el perfil completo")
print("------------------------------------")
```

```python

# Solicitud de la opción
opcion = input("Seleccione una opción del 1 al 6: ")

# Opción 1: agregar un dato nuevo
if opcion == "1":

    clave_nueva = input("Ingrese el nombre de la nueva clave: ")
    valor_nuevo = input("Ingrese el valor asociado a la clave: ")

    perfil[clave_nueva] = valor_nuevo

    print("Dato agregado correctamente.")
    print("Perfil actualizado:")
    print(perfil)

# Opción 2: actualizar un dato existente
elif opcion == "2":

    clave_actualizar = input("Ingrese la clave que desea actualizar: ")

    if clave_actualizar in perfil:

        nuevo_valor = input("Ingrese el nuevo valor: ")
        perfil[clave_actualizar] = nuevo_valor

        print("Dato actualizado correctamente.")
        print("Perfil actualizado:")
        print(perfil)

    else:
        print("La clave ingresada no existe en el perfil.")

# Opción 3: eliminar un dato utilizando del
elif opcion == "3":

    clave_eliminar = input("Ingrese la clave que desea eliminar: ")

    if clave_eliminar in perfil:

        del perfil[clave_eliminar]

        print("Dato eliminado correctamente utilizando del.")
        print("Perfil actualizado:")
        print(perfil)

    else:
        print("La clave ingresada no existe en el perfil.")

# Opción 4: eliminar un dato utilizando pop
elif opcion == "4":

    clave_eliminar = input("Ingrese la clave que desea eliminar: ")

    if clave_eliminar in perfil:

        valor_eliminado = perfil.pop(clave_eliminar)

        print("Dato eliminado correctamente utilizando pop.")
        print("Valor eliminado:", valor_eliminado)
        print("Perfil actualizado:")
        print(perfil)

    else:
        print("La clave ingresada no existe en el perfil.")

# Opción 5: consultar un dato utilizando get
elif opcion == "5":

    clave_consultar = input("Ingrese la clave que desea consultar: ")

    valor_consultado = perfil.get(clave_consultar, "No disponible")

    print("Resultado de la consulta:")
    print(valor_consultado)

# Opción 6: mostrar el perfil completo
elif opcion == "6":

    print("Perfil completo:")
    print(perfil)

# Opción inválida
else:
    print("Opción inválida. Debe seleccionar un número del 1 al 6.")
```
