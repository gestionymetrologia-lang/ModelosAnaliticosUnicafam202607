TALLER ACTUALIZACIÓN DE DATOS

```python
# DICCIONARIO CON DATOS BÁSICOS
estudiante={
    "Nombres":"Manuel Jacinto",
    "Apellidos": "García",
    "Ciudad": "Bogotá",
    "Hijos": "Sí",
    "Carrera": "Ingeniería Eléctrica",
    }
```

```python
# CONSTRUCCIÓN DEL MENÚ DEL USUARIO
print("----BIENVENIDO A TU PROGRAMA DE ACUTALIZACIÓN DE DATOS----")
print("Para comenzar elige la opción deseada")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con pop")
print("5. Consultar un dato")
print("6. Imprimir diccionario completo")
selección=int(input("Selecciona la opción deseada:"))
```

```python
# AGREGAR UN DATO NUEVO
if selección==1:
  clave=input("Ingrese la clave deseada:")
  valor=input("ingrese el valor deseado:")

# AGREGAR UN ELEMENTO NUEVO

  estudiante[clave]=valor
  print("El valor se agregó existosamante")
  print(estudiante)

# ACTUALIZAR DATO

elif selección==2:
  clave=input("Ingrese la clave a actualizar")
  estudiante[clave]=input("Ingrese valor nuevo")
  print("El dato fue actualizado con éxito")
  print(estudiante)

# ELIMINAR DATO CON DEL

elif selección==3:
  clave=input("Ingrese la clave a eliminar")
  del estudiante[clave]
  print("El dato fue eliminado exitosamente")
  print(estudiante)

# ELIMINAR DATO CON POP

elif selección==4:
  clave=input("Ingrese clave a eliminar")
  estudiante.pop(clave)
  print("El dato fue eliminado existosamente")
  print(estudiante)

# CONSULTAR DATO

elif selección==5:
  clave=input("Ingrese clave a consultar")
  valor=estudiante.get(clave, "No disponible")
  print("{clave}:{valor}")

# IMPRIMIR DICCIONARIO COMPLETO

elif selección==6:
  print("Perfil Completo")
  print(estudiante)

else:
  print("Opción Inválida")



```
