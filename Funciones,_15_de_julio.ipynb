```python
# Genera una calculadora de operaciones
def operaciones (a,b):
  suma=a+b
  resta=a-b
  multiplicación=a*b
  división=a/b
  return suma,resta,multiplicación,división
result_suma, result_resta, result_multiplicación, result_división= operaciones(47,35)
print(f"Suma: {result_suma}, Resta: {result_resta}, Multiplicación: {result_multiplicación}, División: {result_división}")
```

```python
# FUNCIÓN PARA DETERMINAR SI UNA PERSONA ES MAYOR O MENOR DE EDAD

def determinar_mayoria_edad(edad):
    if edad >= 18:
        return "La persona es mayor de edad."
    else:
        return "La persona es menor de edad."


# SOLICITAR LA EDAD AL USUARIO
edad_persona = int(input("Ingrese la edad de la persona: "))

# LLAMAR LA FUNCIÓN E IMPRIMIR EL RESULTADO

resultado = determinar_mayoria_edad(edad_persona)
print(resultado)
```

```python
# FUNCIÓN PARA CLASIFICAR EL RANGO DE UNA NOTA OBTENIDA

def clasificar_nota(nota):
    if nota < 0 or nota > 100:
        return "La nota ingresada no es válida."
    elif nota >= 90:
        return "Excelente"
    elif nota >= 70:
        return "Bueno"
    elif nota >= 50:
        return "Suficiente"
    else:
        return "Insuficiente"

# SOLICITAR LA NOTA
nota_estudiante = int(input("Ingrese una nota entre 0 y 100: "))

# LLAMAR LA FUNCIÓN Y MOSTRAR EL RESULTADO
resultado = clasificar_nota(nota_estudiante)

print("Clasificación:", resultado)
```

```python
# FUNCIÓN PARA VERIFICAR UNA CONTRASEÑA

def verificar_contraseña(contraseña):
    if contraseña == "secreta123":
        return "La contraseña es correcta."
    else:
        return "La contraseña es incorrecta."

contraseña_usuario = input("Ingrese la contraseña: ")

resultado = verificar_contraseña(contraseña_usuario)

print(resultado)
```

```python
# FUNCIÓN PARA CALCULAR Y CLASIFICAR EL IMC (ÍNDICE DE MASA CORPORAL)

def imc(peso, altura):
    altura_metros = altura / 100
    resultado_imc = peso / (altura_metros * altura_metros)

    if resultado_imc < 18.5:
        return "Bajo peso"
    elif resultado_imc < 25:
        return "Normal"
    elif resultado_imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"


peso_persona = int(input("Ingrese el peso en kilogramos: "))
altura_persona = int(input("Ingrese la altura en centímetros: "))

rango = imc(peso_persona, altura_persona)

print("Clasificación IMC:", rango)
```
